**Key Findings**

These are a few key findings from the information above:


1. The dataset shows property listings in Jabodetabek from the web 'Rumah123'
2. Almost half of the data in the column 'building_orientation' is missing
3. The latitude and longitude range is very small, so it likely doesn't matter significantly
4. The 'property_type' column all have the same value
5. Information provided on the 'address' column is the same as column 'district' and 'city'
6. 'year_built' have an impossible value 2052
7. The column 'year_built' and 'building_age' doesn't match so it is inconsistent if we use both


```python
# Dropping irrelevant columns
# 'url', 'ads_id', and 'title' is not relevant since it is only additional identity
# 'address' is already represented by 'district' and 'city', but since 'city'
# column is already specific enough, so i decided to also drop 'district'
# 'property_type' have all the same value
# 'lat' and 'long' is dropped because the value range is very small
# 'year_built' is inconsistent and includes impossible values, so using 'building_age' is better

df = df.drop('url', axis=1)
df = df.drop('ads_id', axis=1)
df = df.drop('title', axis=1)
df = df.drop('address', axis=1)
df = df.drop('district', axis=1)
df = df.drop('property_type', axis=1)
df = df.drop('lat', axis=1)
df = df.drop('long', axis=1)
df = df.drop('year_built', axis=1)
df.head()
```

```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]


# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)


null_df
```

```python
# Display unique values for each column
for col in df.columns:
  print(f"Unique values for column '{col}':")
  print(df[col].unique())
  print("-" * 20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only related numerical columns for histogram
selected_cols = ['bedrooms', 'bathrooms', 'land_size_m2', 'building_size_m2', 'floors', 'building_age']

# Create histograms for each numerical column
for col in selected_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

From this histogram, we know that the data is positively skewed, meaning for handling missing values, we shoulde use either median/mean.

```python
# Fill null values in specified numerical columns with the median
cols_to_impute_median = ['bedrooms', 'bathrooms', 'land_size_m2', 'building_size_m2', 'building_age', 'floors']
for col in cols_to_impute_median:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
# For the 'certificate' column, the null values will be replaced by 'lainnya (ppjb,girik,adat,dll)', since it basically represents unknown values

df['certificate'].fillna('lainnya (ppjb,girik,adat,dll)', inplace=True)
```

```python
# Since 'building_orientation' is a categorical column, so I change it to 'Unknown'
df['building_orientation'].fillna('Unknown', inplace=True)
```

```python
# For null values in 'furnishing', if 'property_condition' value is
# 'semi furnished' or 'unfurnished', replace the current value with that value.
# But if it is any other values, replace null values with 'Unknown'.
def fill_furnishing(row):
    if pd.isnull(row['furnishing']):
        if row['property_condition'] in ['semi furnished', 'unfurnished']:
            return row['property_condition']
        else:
            return 'Unknown'
    return row['furnishing']

df['furnishing'] = df.apply(fill_furnishing, axis=1)

# For the null values in 'property_condition', if the 'furnishing' value is 'baru', then replace it with that value.
# If it is others, replace with 'Unknown'.
df['property_condition'] = df.apply(lambda row: row['furnishing'] if pd.isnull(row['property_condition']) and row['furnishing'] == 'baru' else ('Unknown' if pd.isnull(row['property_condition']) else row['property_condition']), axis=1)

# Replace every value in 'property_condition' that says 'semi furnished' or 'unfurnished' to 'Unknown'.
df['property_condition'] = df['property_condition'].replace(['semi furnished', 'unfurnished'], 'Unknown')

# In 'furnishing', change values that says 'baru' with 'Unknown'.
df['furnishing'] = df['furnishing'].replace('baru', 'Unknown')

# This whole process is to make the values of these two column more consistent, which also helps us do feature engineering on the next step
```

```python
# Display null count for each column
print("Null values per column:")
print(df.isnull().sum())
```

```python
#Binning 'building_age' column
import numpy as np

# Define the bins and labels
bins = [-np.inf, 5, 10, 15, 20, 25, np.inf]
labels = ['0-5 years', '6-10 years', '11-15 years', '16-20 years', '21-25 years', '>25 years']

# Replace the 'building_age' column with the new binned categories
df['building_age'] = pd.cut(df['building_age'], bins=bins, labels=labels, right=True)
```

```python
# Feature engineering from total scoring of 'furnishing', 'property_condition', and 'building_age'

# Define scoring dictionaries for 'furnishing', 'property_condition', and 'building_age'
furnishing_scores = {
    'Unknown': 0,
    'unfurnished': 0,
    'semi furnished': 1,
    'furnished': 2
}

property_condition_scores = {
    'Unknown': 0,
    'butuh renovasi': 0,
    'bagus': 1,
    'bagus sekali': 2,
    'sudah renovasi': 2,
    'baru': 3
}

building_age_scores = {
    '0-5 years': 5,
    '6-10 years': 4,
    '11-15 years': 3,
    '16-20 years': 2,
    '21-25 years': 1,
    '>25 years': 0
}

# Map the scores to the columns and create new temporary columns
df['furnishing_score'] = df['furnishing'].map(furnishing_scores)
df['property_condition_score'] = df['property_condition'].map(property_condition_scores)
df['building_age_score'] = df['building_age'].map(building_age_scores).astype(int)

# Create the 'house_quality' feature by summing the scores
df['house_quality'] = df['furnishing_score'] + df['property_condition_score'] + df['building_age_score']

# Display the first few rows with the new feature
display(df[['furnishing', 'property_condition', 'building_age', 'furnishing_score', 'property_condition_score', 'building_age_score', 'house_quality']].head())

# Drop the temporary score columns
df = df.drop(['furnishing_score', 'property_condition_score', 'building_age_score'], axis=1)
```

This way, we can determine the quality score of the houses based on their condition, furnishing, and building age, which are the 3 most important thing people observe when buying a certain property/house.

```python
# Count the number of facilities in each row
df['facilities_count'] = df['facilities'].apply(lambda x: len(x.split(',')))

# Define the bins and labels based on facilities count
bins = [0, 3, 6, np.inf]
labels = ['1-3 facilities', '4-6 facilities', '>7 facilities']

# Replace the 'facilities' column with the new binned categories
df['facilities'] = pd.cut(df['facilities_count'], bins=bins, labels=labels, right=True)

# Drop the temporary 'facilities_count' column
df = df.drop('facilities_count', axis=1)
```
